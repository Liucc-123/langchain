## 2.1 ä½¿ç”¨ReActæ¨¡å¼

```python
# 1.å¯¼å…¥ç›¸å…³ä¾èµ–
from langchain_openai import ChatOpenAI
from langchain_classic.agents import initialize_agent, AgentType
from langchain_classic.tools import Tool
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_experimental.utilities.python import PythonREPL

# 2. è®¾ç½® TAVILY_API å¯†é’¥
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")

# 3.å®šä¹‰æœç´¢å·¥å…·
search = TavilySearchResults(max_results=3)

search_tool = Tool(
    name="Search",
    func=search.run,
    description="ç”¨äºæœç´¢äº’è”ç½‘ä¸Šçš„ä¿¡æ¯ï¼Œç‰¹åˆ«æ˜¯è‚¡ç¥¨ä»·æ ¼å’Œæ–°é—»"
)

# 4.å®šä¹‰è®¡ç®—å·¥å…·
python_repl = PythonREPL()  # LangChainå°è£…çš„å·¥å…·ç±»å¯ä»¥è¿›è¡Œæ•°å­¦è®¡ç®—

calc_tool = Tool(
    name="Calculator",
    func=python_repl.run,
    description="ç”¨äºæ‰§è¡Œæ•°å­¦è®¡ç®—ï¼Œä¾‹å¦‚è®¡ç®—ç™¾åˆ†æ¯”å˜åŒ–"
)

# 5. å®šä¹‰LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)

# 6. åˆ›å»ºAgentExecutoræ‰§è¡Œå™¨å¯¹è±¡
agent_executor = initialize_agent(
    tools=[search_tool, calc_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# 7. æµ‹è¯•è‚¡ç¥¨ä»·æ ¼æŸ¥è¯¢
query = "å°ç±³å…¬å¸å½“å‰è‚¡ä»·æ˜¯å¤šå°‘ï¼Ÿç›¸è¾ƒå»å¹´ä¸Šæ¶¨äº†è¿˜æ˜¯è·Œäº†ï¼Ÿå¦‚æœæ¶¨äº†ï¼Œæ¶¨äº†å¤šå°‘ï¼Ÿå¦‚æœè·Œäº†ï¼Œè·Œäº†å¤šå°‘ï¼Ÿ"
result = agent_executor.invoke(query)
print(f"æŸ¥è¯¢ç»“æœ: {result}")
```

## 2.2 ä½¿ç”¨FUNCTION_CALLæ¨¡å¼


```python
from langchain_classic.agents import AgentExecutor
from langchain_core.prompts import ChatPromptTemplate
import os, dotenv
# 1.å¯¼å…¥ç›¸å…³ä¾èµ–
from langchain_openai import ChatOpenAI
from langchain_classic.agents import create_openai_functions_agent
from langchain_classic.tools import Tool
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_experimental.utilities.python import PythonREPL


# 2. è®¾ç½® TAVILY_API å¯†é’¥
dotenv.load_dotenv()
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")

# 3.å®šä¹‰æœç´¢å·¥å…·
search = TavilySearchResults(max_results=3)

search_tool = Tool(
    name="Search",
    func=search.run,
    description="ç”¨äºæœç´¢äº’è”ç½‘ä¸Šçš„ä¿¡æ¯ï¼Œç‰¹åˆ«æ˜¯è‚¡ç¥¨ä»·æ ¼å’Œæ–°é—»"
)

# 4.å®šä¹‰è®¡ç®—å·¥å…·
python_repl = PythonREPL()  # LangChainå°è£…çš„å·¥å…·ç±»å¯ä»¥è¿›è¡Œæ•°å­¦è®¡ç®—

calc_tool = Tool(
    name="Calculator",
    func=python_repl.run,
    description="ç”¨äºæ‰§è¡Œæ•°å­¦è®¡ç®—ï¼Œä¾‹å¦‚è®¡ç®—ç™¾åˆ†æ¯”å˜åŒ–"
)

# 5.å®šä¹‰æç¤ºè¯ï¼ˆcreate_openai_functions_agentæ–¹å¼ä¸‹å¿…é¡»æŒ‡å®špromptï¼‰
prompt = ChatPromptTemplate.from_messages([
    ("system", "ä½ æ˜¯ä¸€ä¸ªå¯ä»¥è®¿é—®äº’è”ç½‘ä¿¡æ¯çš„æ™ºèƒ½åŠ©æ‰‹ã€‚"),
    ("human", "{input}"),
    ("system", "{agent_scratchpad}")
])

# 6.å®šä¹‰LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)

# 7.åˆ›å»ºæ–°ç‰ˆAgentï¼ˆæ›¿ä»£ initialize_agentï¼‰
agent = create_openai_functions_agent(
    tools=[search_tool, calc_tool],
    llm=llm,
    prompt=prompt
)

# 8.åˆ›å»ºagentæ‰§è¡Œå™¨
agent_executor = AgentExecutor(
    tools=[search_tool, calc_tool],
    agent=agent,
    verbose=True
)

# 9.æµ‹è¯•è‚¡ç¥¨ä»·æ ¼æŸ¥è¯¢
query = "å°ç±³å…¬å¸å½“å‰è‚¡ä»·æ˜¯å¤šå°‘ï¼Ÿç›¸è¾ƒå»å¹´ä¸Šæ¶¨äº†è¿˜æ˜¯è·Œäº†ï¼Ÿå¦‚æœæ¶¨äº†ï¼Œæ¶¨äº†å¤šå°‘ï¼Ÿå¦‚æœè·Œäº†ï¼Œè·Œäº†å¤šå°‘ï¼Ÿ"
result = agent_executor.invoke(input={"input": query})
print(f"æŸ¥è¯¢ç»“æœ: {result}")
```

ä¸¾ä¾‹ï¼šè®¡ç®—3çš„å¹³æ–¹ï¼ŒAgentè‡ªåŠ¨è°ƒç”¨å·¥å…·å®Œæˆ

```python
from langchain_classic.agents import initialize_agent, AgentType, Tool
from langchain_openai import ChatOpenAI


# 1. å®šä¹‰å·¥å…· - è®¡ç®—å™¨ï¼ˆè¦æ±‚å­—ç¬¦ä¸²è¾“å…¥ï¼‰
def simple_calculator(expression: str) -> str:
    """
    åŸºç¡€æ•°å­¦è®¡ç®—å·¥å…·ï¼Œæ”¯æŒåŠ å‡ä¹˜é™¤å’Œå¹‚è¿ç®—
    å‚æ•°:
        expression: æ•°å­¦è¡¨è¾¾å¼å­—ç¬¦ä¸²ï¼Œå¦‚ "3+5" æˆ– "2**3"
    è¿”å›:
        è®¡ç®—ç»“æœå­—ç¬¦ä¸²æˆ–é”™è¯¯ä¿¡æ¯
    """
    print(f"\n[å·¥å…·è°ƒç”¨] è®¡ç®—è¡¨è¾¾å¼: {expression}")

    print("åªå› ä¸ºåœ¨äººç¾¤ä¸­å¤šçœ‹äº†ä½ ä¸€çœ¼ï¼Œç¡®è®¤ä¸‹ä½ è°ƒç”¨äº†æˆ‘^_^")
    return str(eval(expression))


# 2. åˆ›å»ºå·¥å…·å¯¹è±¡
math_calculator_tool = Tool(
    name="Math_Calculator",  # å·¥å…·åç§°ï¼ˆAgentå°†æ ¹æ®åç§°é€‰æ‹©å·¥å…·ï¼‰
    func=simple_calculator,  # å·¥å…·è°ƒç”¨çš„å‡½æ•°
    description="ç”¨äºæ•°å­¦è®¡ç®—ï¼Œè¾“å…¥å¿…é¡»æ˜¯çº¯æ•°å­¦è¡¨è¾¾å¼ï¼ˆå¦‚'3+5'æˆ–'3**2'è¡¨ç¤ºå¹³æ–¹ï¼‰ã€‚ä¸æ”¯æŒå­—æ¯æˆ–ç‰¹æ®Šç¬¦å·"  # å…³é”®ï¼šæ˜ç¡®è¾“å…¥æ ¼å¼è¦æ±‚
)

# 3. åˆå§‹åŒ–å¤§æ¨¡å‹
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)

# 4. åˆå§‹åŒ–AgentExecutorï¼ˆä½¿ç”¨é›¶æ ·æœ¬Reactæ¨¡å¼ã€å¢åŠ è¶…æ—¶è®¾ç½®ï¼‰
agent_executor = initialize_agent(
    tools=[math_calculator_tool],  # å¯ç”¨çš„å·¥å…·åˆ—è¡¨
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,  # ç®€å•æŒ‡ä»¤æ¨¡å¼
    verbose=True  # å…³é”®å‚æ•°ï¼åœ¨æ§åˆ¶å°æ˜¾ç¤ºè¯¦ç»†çš„æ¨ç†è¿‡ç¨‹
)

# 5. æµ‹è¯•å·¥å…·è°ƒç”¨ï¼ˆæ·»åŠ å¼‚å¸¸æ•è·ï¼‰
print("=== æµ‹è¯•ï¼šæ­£å¸¸å·¥å…·è°ƒç”¨ ===")
response = agent_executor.invoke("è®¡ç®—3çš„å¹³æ–¹")  # å‘Agentæé—®
print("æœ€ç»ˆç­”æ¡ˆ:", response)
```
